```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca gambar
img = cv2.imread('Output/3.jpg')

# Mengonversi gambar menjadi grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Mendeteksi sudut menggunakan algoritma Shi-Tomasi
corners = cv2.goodFeaturesToTrack(gray_img, 100, 0.01, 10)

# Konversi hasil deteksi ke tipe integer menggunakan np.intp
corners = np.intp(corners)

# Menggambar lingkaran di setiap titik sudut
for i in corners:
    x, y = i.ravel()
    cv2.circle(img, (x, y), 3, (255, 0, 0), -1)

# Mengonversi gambar dari BGR ke RGB agar tampilannya benar di matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Menampilkan gambar hasil dengan matplotlib
plt.imshow(img_rgb)
plt.axis('off')  # Menghilangkan sumbu
plt.show()
```
